## Изменение данных в таблицах

```python
file = "for_lin_gui_ba_fa"
df = pd.read_csv(f"myauthapp/accounts/data/{file}.csv")
df
for c in df.columns[1:]:
    df[c] = df[c].str.replace("Жэнь", "Жень")
    df[c] = df[c].str.replace("Чун", "Чонг")
    df[c] = df[c].str.replace("цзяо", "цяо")
df
# df.to_csv(f"myauthapp/accounts/data/{file}.csv", index=False)
```

```python
# Уменьшение длины ряда временных периодов 
times = ["01.30 - 03.00", "03.00 - 4.30", "04.30 - 5.00", "05.00 - 7.00", "07.00 - 7.30", "13.30 - 15.00", "19.00 - 19.30"]

rows = ""
for t in times:
    if len(rows) == 0:
        rows += f"{t}<br>"
    else:
        endp = int(rows.rstrip("<br>").split(" - ")[-1].split(".")[0])
        nextp = int(t.split(" - ")[0].split(".")[0])
        if endp == nextp:
            rows = rows.replace(rows.rstrip("<br>").split(" - ")[-1], t.split(" - ")[1])
        else:
            rows += f"{t}<br>"
            
print(rows)
```
