```python
df = pd.read_csv("myauthapp/accounts/data/tai_yan_ba_fa/trid.csv")
df[:5]
```

```python
for c in df.columns[2:]:
    df[c] = df[c].str.replace("Жэнь", "Жень")
    df[c] = df[c].str.replace("Чун", "Чонг")
    df[c] = df[c].str.replace("цзяо", "цяо")
df
```

```python
df.to_csv("myauthapp/accounts/data/tai_yan_ba_fa/trid.csv", index=False)
```
